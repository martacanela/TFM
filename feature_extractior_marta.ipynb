```python
from radiomics import featureextractor
import os
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import SimpleITK as sitk
```

```python
def radiomic_features_extraction(dataDir, params):
    df = pd.DataFrame()

    # Loop through each subdirectory in dataDir
    for case_id in os.listdir(dataDir):
        case_path = os.path.join(dataDir, case_id)
        
        # Check if it is a directory
        if os.path.isdir(case_path):
            # Define paths for the image and mask files based on the case directory name
            imagePath = os.path.join(case_path, "Image.nii")
            labelPath = os.path.join(case_path, "Mask.nii")

            print(f"Processing case: {case_id}")
            print(f"Image path: {imagePath}")
            print(f"Label path: {labelPath}")

            # Initialize the feature extractor with parameters
            extractor = featureextractor.RadiomicsFeatureExtractor(params)
            extractor.enableAllFeatures()

            try:
                # Extract features
                result = extractor.execute(imagePath, labelPath)

                # Append the features to the DataFrame, using case_id as an identifier
                feature_series = pd.Series(result)  # Convert result dictionary to Series
                feature_series['case_id'] = case_id  # Add case_id to the Series
                df = df.append(feature_series, ignore_index=True)  # Append Series as a new row
                
            except Exception as e:
                print(f"Error processing case {case_id}: {e}")

    # Drop diagnostic columns if they exist
    for col in df.columns:
        if "diagnostics" in col:
            df.drop([col], axis=1, inplace=True)

    return df 

dataDir = "E:/TFM Marta/Imatges_Marta_nifti/"
params = os.path.join(os.getcwd(), "E:/TFM Marta/Codis/Judith Osuna/param1.yaml")

df_worked = radiomic_features_extraction(dataDir, params)
```
